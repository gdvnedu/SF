```python
import pandas as pd
import re

log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

def alter_user_id(user_id):
    pattern = re.compile('Запись пользователя № - user_\d+')
    if pattern.match(user_id):
        user = user_id[user_id.find('user_'):]
        return user
    else:
        return ''

log['user_id'] = log.user_id.apply(alter_user_id)

display(log)
```

```python
import pandas as pd
import re


def alter_user_id(user_id):
    pattern = re.compile('Запись пользователя № - user_\d+')
    if pattern.match(user_id):
        user = user_id[user_id.find('user_'):]
        return user
    else:
        return ''
    
def alter_time(time):
    pattern = re.compile('\[.+')
    if pattern.match(str(time)):
        time = time[time.find('[')+1:]
        return time 

log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

log = log[log.user_id.str.match("Запись пользователя", na=False)]
log['user_id'] = log.user_id.apply(alter_user_id)
log['time'] = log.time.apply(alter_time)

display(log)
```
