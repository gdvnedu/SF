```python
import pandas as pd
ratings = pd.read_csv('ratings_example.txt', sep = '\t')
movies = pd.read_csv('movies_example.txt', sep = '\t')
movies.drop_duplicates(subset = 'movieId', keep = 'first', inplace = True)
ratings.merge(movies, how = 'left', on = 'movieId')
```
